# Matplotlibについて

```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
matplotlib.style.use("ggplot")
```

## Matplotlib概要

>MATLABスタイル

```python
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y)
plt.title('MATLAB-style')

plt.show()
```

>オブジェクト指向スタイル

```python
x = [1, 2, 3]
y = [2, 4, 9]

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("OPP-style")
plt.show()
```

## グラフの種類と出力方法

>折れ線グラフ

```python
fig, ax = plt.subplots()
# データが細かくなると曲線になる。
x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 5]
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

>棒グラフ

```python
fig, ax = plt.subplots()
x = [1, 2, 3]
y = [10, 2, 3]
ax.bar(x, y)
plt.show()
```

```python
# labelをつける
fig, ax = plt.subplots()
x = [1, 2, 3]
y = [10, 2, 3]
labels = ["spam", "ham", "egg"]
# barhで横向きの棒グラフ
ax.barh(x, y, tick_label=labels)
plt.show()
```

```python
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]

width = 0.4
ax.bar(x, y1, width=width, tick_label=labels, label="y1")
x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label="y2")

ax.legend()
plt.show()
```

```python
# 積み上げグラフ
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ["spam", "ham", "egg"]

y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]
ax.bar(x, y_total, tick_label=labels, label="y1")
ax.bar(x, y2, label="y2")
ax.legend()
plt.show()
```

>散布図

```python
fig, ax = plt.subplots()
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)
ax.scatter(x, y)
plt.show()
```

```python
fig, ax = plt.subplots()
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)
ax.scatter(x[0:10], y[0:10], marker="v", label="triangle down")
ax.scatter(x[10:20], y[10:20], marker="^", label="triangle up")
ax.scatter(x[20:30], y[20:30], marker="s", label="square")
ax.scatter(x[30:40], y[30:40], marker="*", label="star")
ax.scatter(x[40:50], y[40:50], marker="x", label="x")
ax.legend()

plt.show()
```

>ヒストグラム

```python
np.random.seed(123)
mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(x, bins=20)

plt.show()
```

```python
# 横向き
np.random.seed(123)
mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()
n, bins, patches = ax.hist(x, orientation="horizontal")

plt.show()
```

```python
# ヒストグラムを並べる
np.random.seed(123)
mu = 100
x0 = np.random.normal(mu, 20, 1000)
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()
labels = ["x0", "x1", "x2"]
ax.hist((x0, x1, x2), label=labels)
ax.legend()
plt.show()
```

```python
# ヒストグラムを積み上げる
np.random.seed(123)
mu = 100
x0 = np.random.normal(mu, 20, 1000)
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()
labels = ["x0", "x1", "x2"]
ax.hist((x0, x1, x2), label=labels, stacked=True)
ax.legend()
plt.show()
```

>箱ひげ図

```python
np.random.seed(123)
x0 = np.random.normal(0, 10, 500)
x1 = np.random.normal(0, 15, 500)
x2 = np.random.normal(0, 20, 500)

fig, ax = plt.subplots()
labels = ["x0", "x1", "x2"]
ax.boxplot((x0, x1, x2), labels=labels)
plt.show()
```

```python
# 横向き
fig, ax = plt.subplots()
labels = ["x0", "x1", "x2"]
ax.boxplot((x0, x1, x2), labels=labels, vert=False)
plt.show()
```

>円グラフ

```python
labels = ["spam", "ham", "egg"]
x = [10, 3, 1]

fig, ax = plt.subplots()
ax.pie(x, labels=labels)
ax.axis("equal")
plt.show()
```

```python
# 向きの調整
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False)
ax.axis("equal")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct="%1.2f%%")
ax.axis("equal")
plt.show()
```

```python
explode = [0, 0.2, 0]

fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct="%1.2f%%", explode=explode)
ax.axis("equal")
plt.show()
```

>複数のグラフを組み合わせる

```python
fig, ax = plt.subplots()

x1 = [1, 2, 3]
y1 = [5, 2, 3]
x2 = [1, 2, 3, 4]
y2 = [8, 5, 4, 6]
ax.bar(x1, y1, label="y1")
ax.plot(x2, y2, label="y2")
ax.legend()
plt.show()
```

```python
# ヒストグラムと近似戦をプロット
np.random.seed(123)
x = np.random.randn(1000)

fig, ax = plt.subplots()
counts, edges, patches = ax.hist(x, bins=25)
x_fit = (edges[:-1] + edges[1:]) / 2
y = 1000 * np.diff(edges) * np.exp(-x_fit**2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)
plt.show()
```

## スタイル

>色の設定

```python
fig, ax = plt.subplots()
ax.plot([1, 3], [3, 1], label="aqua", color="aqua")
ax.plot([1, 3], [1, 3], label="#0000FF", color="#0000FF")
ax.plot([1, 3], [2, 2], label="(0.1, 0.2, 0.5, 0.3)", color=(0.1, 0.2, 0.5, 0.3))
ax.legend()
plt.show()
```

```python
# 塗りつぶしと枠線
fig, ax = plt.subplots()

ax.bar([1], [3], color="aqua")
ax.bar([2], [4], color="aqua", edgecolor="black")
plt.show()
```

>線のスタイル

```python
fig, ax = plt.subplots()
ax.plot([1, 3], [3, 1], linewidth=5.5, label="5.5")
ax.plot([1, 3], [1, 3], linewidth=10, label="10")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 3], [3, 1], label="dashed", linestyle="--")
ax.plot([1, 3], [1, 3], label="dashdot", linestyle="-.")
ax.plot([1, 3], [2, 2], label="dotted", linestyle=":")
ax.legend()
plt.show()
```

>フォント

```python
fig, ax = plt.subplots()
ax.set_xlabel("xlabel", family="fantasy", size=20, weight="bold")
ax.set_ylabel("ylabel", family="cursive", size=40, weight="light")

fontdict = {
    "family": "monospace",
    "size": 25,
    "weight": "heavy"
}
ax.set_title("graph title", fontdict=fontdict)
plt.show()
```

>テキスト描画

```python
fig, ax = plt.subplots()
ax.text(0.2, 0.4, "text", size=20)
plt.show()
```
